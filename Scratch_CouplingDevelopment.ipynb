### Attempt #1: Pass wpthlp from ctsm and prescribe that at the surface

```python
dataDir = '/Users/mdfowler/Documents/Analysis/CLASP/SCAM_runs/TempFiles/'

noCouplerFile = dataDir+'FSCAM.T42_T42.testCESM_sanitySandbox.claspIOP.cam.h0.2015-06-01-84585.nc'
camFile = dataDir+'FSCAM.T42_T42.testPassingCLUBB_all4FldsInInterfaceMod.001.cam.h0.2015-06-01-84585.nc'
clmFile = dataDir+'FSCAM.T42_T42.testPassingCLUBB_all4FldsInInterfaceMod.001.clm2.h0.2015-06-01-84585.nc'

with xr.open_dataset(noCouplerFile, decode_times=True) as noCplDS:
    noCplDS['time'] = noCplDS.indexes['time'].to_datetimeindex()
    
with xr.open_dataset(camFile, decode_times=True) as camDS:
    camDS['time'] = camDS.indexes['time'].to_datetimeindex()
    
with xr.open_dataset(clmFile, decode_times=True) as clmDS:
    clmDS['time'] = clmDS.indexes['time'].to_datetimeindex()
```

```python
fig, ax1 = plt.subplots(1,1, figsize=(8, 5))

ax1.plot(noCplDS['time'].values, np.squeeze(noCplDS['WPTHLP_CLUBB'].values)[:,-1], label='CAM: noCpl')
ax1.plot(camDS['time'].values,   np.squeeze(camDS['WPTHLP_CLUBB'].values)[:,-1], label='CAM: cpl')
# ax1.plot(camDS['time'].values,   np.squeeze(clmDS['WPTHLP_CLUBB'].values)[:], label='CLM')

ax1.legend(fontsize=12)
ax1.set_title('WPTHLP_CLUBB', fontsize=14)
ax1.set_ylabel(camDS['WPTHLP_CLUBB'].units)
```
